```python
df = pd.DataFrame()

for i in range(6):
    response_API = requests.get(f'https://api.themoviedb.org/3/movie/55{i}?api_key=e31a75ca2478a169b405cf6ec5313235')
    dataframe = pd.DataFrame.from_dict(response_API.json(), orient = 'index')
    dft = dataframe.transpose()
    df = df.append(dft, ignore_index = True)
    
df
```

```python
#2.3 Transform
#make backup
df_bck = df.copy() 
#filter cols
df_bck.columns
select_col = ['budget', 'genres', 'id', 'imdb_id', 'original_title', 'release_date', 'revenue','runtime']
df_bck = df_bck[select_col]
df_bck
```

```python
#create seperate dataframe to get genre list
df_bck_genre = df_bck['genres']
df_bck_genre


## create func. to add list of genre values from genre column

def get_genre(i):
    a = df_bck_genre[i]
    my_list = []
    for d in a:
        name = d.get('name')
        my_list.append(name)

    return my_list

#Add genre to Movies table (df_bck)

df_bck['genre_name'] = ""

for i in range(6):
    df_bck['genre_name'][i] =get_genre(i)
    

df_bck

```

```python
#Drop genre column
df_bck1 = df_bck.copy() ### 
df_bck1.drop('genres',axis=1,inplace=True)

#Add genre frequency column to df_bck1

Total_genres =['Drama','Thriller','Comedy','Action','Adventure','Romance','History','Crime']

def get_freq(i):
    y = df_bck1['genre_name'][i]
    z =[]
    for elem in Total_genres:
        if elem in y:
            z.append("1")
        else:
            z.append("0")
    return z

df_bck1['genre_freq']=""
for i in range(6):
    df_bck1['genre_freq'][i] =get_freq(i)
    
df_bck1


```

```python
#split genre_freq into columns

split_df = pd.DataFrame(df_bck1['genre_freq'].tolist(), columns=['Drama','Thriller','Comedy','Action','Adventure','Romance','History','Crime'])
# display the resulting df
df_bck1 = pd.concat([df_bck1, split_df], axis=1)
#split_df
df_bck1
```

```python
#Make Movies dataframe with only required value
movies_df = df_bck1.copy()
movies_df.drop('genre_freq',axis=1, inplace=True)
movies_df.drop('genre_name',axis=1, inplace=True)
movies_df
```

```python
#create seperate genre table
#make copy of orignal dataframe df_genre
df_g = df_bck.copy()
Genre_df = pd.DataFrame(df_g["genres"].explode().tolist())
Genre_df= Genre_df.drop_duplicates()
Genre_df
```

```python
#Transformation 2
# Expand date column

datetimes_df = pd.DataFrame(df_bck1["release_date"])

#convert column to datetime
datetimes_df["release_date"] = pd.to_datetime(datetimes_df["release_date"])
datetimes_df['Id'] = ['550','551','552','553','554','555']

datetimes_df['r_day'] = datetimes_df['release_date'].dt.day
datetimes_df['r_month'] = datetimes_df['release_date'].dt.month
datetimes_df['r_year'] = datetimes_df['release_date'].dt.year
datetimes_df['day_of_week'] = datetimes_df['release_date'].dt.strftime("%A")
datetimes_df = datetimes_df.iloc[:,[1,0,2,3,4,5]]
datetimes_df
```

```python
#Load
Movies_df_csv = movies_df.to_csv('Movies.csv', index = False)
Genre_df_csv = Genre_df.to_csv('Genre.csv', index = False)
datetimes_df_csv = datetimes_df.to_csv('datetimes_df.csv', index = False)
```
